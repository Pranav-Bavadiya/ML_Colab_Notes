## 3 ways to call api

#1 kaggle data
it has it's own library to import data


```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python
import pandas as pd
import requests
```

```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]

# Set the path to the file you'd like to load
file_path = "hollywood_2025_hype_sentiment.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "kanchana1990/hollywood-2025-media-hype-and-sentiment",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

#2 by request
direct get request to api


```python
headers = {
    'Authorization': 'Bearer YOUR_HEADER_HERE' # Replace YOUR_HEADER_HERE with your actual authorization token or other headers
}
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1', headers=headers)
```

```python
temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

for more than one page

```python
for i in range(1,5):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True)
```
